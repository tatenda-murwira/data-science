```python

import pandas as pd

# Path of the file to read
iowa_file_path = '../input/home-data-for-ml-course/train.csv'

home_data = pd.read_csv(iowa_file_path)

# Setting up code checking
from learntools.core import binder
binder.bind(globals())
from learntools.machine_learning.ex3 import *

print("Setup Complete")

# printing the list of columns in the dataset to find the name of the prediction target
home_data.columns

y = home_data.SalePrice


# Create the list of features below
feature_names = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']

# Selecting data corresponding to features in feature_names
X = home_data[feature_names]


from sklearn.tree import DecisionTreeRegressor
iowa_model = DecisionTreeRegressor(random_state=1)

# Fit the model
iowa_model.fit(X,y)

predictions = iowa_model.predict(X)
print(predictions)

```

## Step 1: Specify Prediction Target
The target variable y was defined by selecting the sales price column from the dataset. To ensure accuracy, the column headers were audited to isolate the specific price variable for supervised learning.

## Step 2: Create X
A feature set X was constructed using a subset of the original data. This involved defining a list of specific predictive features: LotArea, YearBuilt, 1stFlrSF, 2ndFlrSF, FullBath, BedroomAbvGr, and TotRmsAbvGrd. These variables were then used to create the final DataFrame for model training.

## Step 3: Specify and Fit Model
A DecisionTreeRegressor was imported from sklearn.tree and initialized as iowa_model. The model was then fitted using the feature DataFrame X and the target vector y, allowing the algorithm to learn the relationship between home attributes and sales prices.

## Step 4: Make Predictions
Using the fitted iowa_model, the predict function was applied to the feature set X. The resulting estimates were stored in a variable called predictions, representing the model's initial output for the training data.
