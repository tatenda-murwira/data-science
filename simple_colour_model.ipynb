```python

import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# We are using RGB values. Format: [Red, Green, Blue]
X_train = np.array([
    [255, 0, 0],    # Pure Red
    [255, 50, 50],  # Light Red
    [0, 0, 255],    # Pure Blue
    [50, 50, 255]   # Light Blue
])


# 0 = Warm (Reds), 1 = Cool (Blues)
y_train_good = np.array([0, 0, 1, 1])


# We use k-Nearest Neighbors. It looks at the single closest example.
model = KNeighborsClassifier(n_neighbors=1)



print("--- PHASE 1: GOOD FUEL ---")
model.fit(X_train, y_train_good)
print("Model trained on accurate data.\n")

# TEST DRIVE
# A strong Purple [150, 0, 200] should be "Cool" (1)
new_color = np.array([[150, 0, 200]])
prediction = model.predict(new_color)
result_name = "Warm" if prediction[0] == 0 else "Cool"

print(f"Testing with Purple color: {new_color[0]}")
print(f"Prediction with GOOD data: {result_name}")


print("\n=================================\n")


print("--- PHASE 2: GARBAGE FUEL ---")
# We completely flip reality. We tell it Reds are Cool (1) and Blues are Warm (0).
y_train_garbage = np.array([1, 1, 0, 0])

model.fit(X_train, y_train_garbage)
print("Model re-trained on garbage data (reality flipped).\n")

# Testing again with the exact same purple color
garbage_prediction = model.predict(new_color)
garbage_result_name = "Warm" if garbage_prediction[0] == 0 else "Cool"

print(f"Testing again with the exact same Purple color: {new_color[0]}")
print(f"Prediction with GARBAGE data: {garbage_result_name.upper()}")
print("(The model confidently gives the wrong answer because the data was bad!)")
```
